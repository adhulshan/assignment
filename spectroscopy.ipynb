```python
import pandas as pd
import  glob as gb
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
path='C:/Users/adhul/Desktop/spectroscopy'
```

```python
files = gb.glob(path + "/*.csv")
```

```python
li = []
for f in files:
    temp_df = pd.read_csv(f)
    li.append(temp_df)
samp = li[0]
df = pd.concat(li,ignore_index='true')
```

```python
samp
```

```python
#picking the median value wavelength
wave = st.median(samp.iloc[:,0])
wave
```

```python
indices = np.where(df.iloc[:,0]==wave)
indices
```

```python
intensity = df.iloc[ df.index[indices], 1]
intensity
```

```python
intensity.index = np.arange( 1,intensity.count()+1)
intensity
```

```python
plt.xlabel("DataFrame")
plt.ylabel("Intensity")
intensity.plot(figsize=(10,7))
plt.show()
```
